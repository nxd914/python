# Inheritance

inheritance.png

In OOP, **inheritance** is the mechanism of deriving (inheriting) the properties from another class.

It allows us to define a child class that inherits all the methods and properties from a parent class.
- **Parent class:** the class being inherited from, a.k.a. *base class* or *super class*.
- **Child class:** the class that inherits from another class, a.k.a. *derived class* or *sub-class*.

Inheritance is one of the most powerful concepts in OOP, which help us to represent real-world relationshipsin programming.

It also improves efficiency by providing code reusability.

Inheritance enables a class to acquire all the members of another class.
 - attributes, methods, or both.

By providing reusability, inheritance makes it easier to create as well as maintain an application.

The class which acquires is known as the *child (or derived) class*.

The one that it acquires from is known as the *superclass (a.k.a., base class or parent class)*.

Four forms of inheritance supported by Python:

 - **Single Inheritance** – A single derived class acquires from on single superclass.
 - **Multi-Level Inheritance** – At least 2 different derived classes acquire from two distinct base classes.
 - **Hierarchical Inheritance** – A number of child classes acquire from one superclass
 - **Multiple Inheritance** – A derived class acquires from several superclasses.

# Hierarchies

- **child (subclass) class**
    * inherits all data and behaviors of parent class

Let's learn inheritance with an example, where we will define two classes: `Bird` and `Owl`.

```python
class Bird:
  def __init__(self):
    print('Bird created')

  def whoAmI(self):
    print('I am a bird.')

  def fly(self):
    print('Birds can fly.')
```

````python
class Bird:
  def __init__(self):
    print('Bird created')

  def whoAmI(self):
    print('I am a bird.')

  def fly(self):
    print('Birds can fly.')

  def swim(self):
    print('Brids can swim')

bird = Bird()
bird.whoAmI()
bird.fly()
bird.swim()
````

The above `Bird` class represents a parent class for all of the bird types. It is a generic class which tells the general properties of birds.

`Owl` is a bird too, so we can think of ti as a child class of class `Bird`.

Let's define this inheritance now:

```python
class Owl(Bird):
  def __init__(self):
    super().__init__()
    print('Owl is created.')

  def whoAmI(self):
    print('I am an owl.')

  def swim(self):
    print('Owls cannot swim.')

  def night_vision(self):
    print('Owls have night vision.')
```

In the example above, we defined class `Owl` which inherits from class `Bird`.

In class definition, the parent class is written inside the parentheses as: `class Owl(Bird)`.
- `Bird` is the parent class and the `Owl` is the child class.
- `Owl` inherits all the properties and methods from its parent.
  * It can use them as they are or it can *override* them.
  * It can also add new attributes and/or methods which are not present in its parent.

In our example, we override `__init__()`, `whoAmI()` and `swim()` methods.

We do not redefine the `fly()` method, which means we use it as it is.

We also have the new method `night_vision()` in the `Owl` class which doesn't exist in its parent.
- Night vision is special for owls, but not common for all the other birds.

**`super()`:**

Within `Owl`'s `__init__()`, we have a special method call, `super()`.

`super()` method lets us to access methods in the parent class.

It returns a temporary object of the super class that then allows us to call that super class's methods.

In the `__init__()` method of the `Owl` class we call `Bird`'s constructor as: `super().__init__()`
- Here, `super()` simply means the `Bird` class.

This is a very common pattern in OOP for the `__init__()` methods; we first call the `__init__()` method of the planet, then we add custom functionality for the child class.

Let's now define `Owl` and create an object to see inheritance in action

````python
class Owl(Bird):
  def __init__(self):
    super().__init__()
    print('Owl is created.')

  def whoAmI(self):
    print('I am an owl.')

  def swim(self):
    print('Owls cannot swim.')

  def night_vision(self):
    print('Owls have night vision.')

owl = Owl()
owl.whoAmI()
owl.fly()
owl.swim()
owl.night_vision()
````

**Remember:** Methods defined in children classes only, are not observed from the parent class.

If you try to call `night_vision()` from the instance `bird` of class `Bird`, you will get an error

````python
bird.night_vision()
````

The last example proves that inheritance **works downwards** (from parent to child) but **not upwards** (from child to parent).

**Comments on inheritance:**

- subclass can have **methods with same name** as superclass
- for an instance of a class, the interpreter looks for a method name in **current class definition**
- if not found, it looks for the method's name **up the hierarchy** (in parent, then grandparent, etc.)
- The first method up the hierarchy that is found with that method name is used.

Although not using a constructor method when inheriting from a super class is possible, sometimes (especially when a child class requires additional inputs) is necessary to include a constructor.

In that case, the constructor has to "call" the parent class' constructor using the builtin `super()`
- `super()` returns a temporary object of the superclass that allows us to access methods of the base class.

`super()` has two major use cases:
- Allows us to avoid using the base class name explicitly
- Working with Multiple Inheritance

# Polymorphism

polymorphism.png

**Polymorphism** is another important concept in OOP.

The meaning of the word is **having many forms**. It refers to the same object (or function) exhibiting different forms and behaviors in different scenarios.
- Means the same function name behaving differently for different classes.

To understand polymorphism we will use three classes: `Bird`, `Owl`, and `Penguin`.
- Parent class: `Bird`.
- `Owl`, `Penguin` are children classes.

Let's redefine `Owl` and `Penguin` classes:

```python
class Owl(Bird):
  def __init__(self):
    print('Owl is created.')

  def whoAmI(self):
    print('I am an owl.')

  def fly(self):
    print('Some owls cannot fly.')

  def swim(self):
    print('Owls cannot swim.')

  def night_vision(self):
    print('Owls have night vision.')
```

We see here that the `Owl`'s constructor does not call `Bird`'s constructor, though the former inherits from the latter.

```python
class Penguin(Bird):
  def __init__(self):
    print('Penguin is created.')

  def whoAmI(self):
    print('I am a Penguin.')

  def fly(self):
    print('Penguins cannot fly.')
```

Again, while `Penguin` inherits from `Bird` it does not call the latter's constructor method.

It also overrides the `whoAmI()` and `fly()` methods, keeping the `swim()` method as it is in the parent class.

We now have two child classes inheriting from the same parent class:


````python
class Owl(Bird):
  def __init__(self):
    print('Owl is created.')

  def whoAmI(self):
    print('I am an owl.')

  def fly(self):
    print('Some owls cannot fly.')

  def swim(self):
    print('Owls cannot swim.')

  def night_vision(self):
    print('Owls have night vision.')

class Penguin(Bird):
  def __init__(self):
    print('Penguin is created.')

  def whoAmI(self):
    print('I am a Penguin.')

  def fly(self):
    print('Penguins cannot fly.')
````

We will define a new function `can_fly()` to see the polymorphism in action. It will be a common function that works on any object which is of type `Bird` or its children.

````python
def can_fly(bird):
  bird.fly()
````

The function is very simple. It has a single parameter named `bird` and calls its method `fly()`.
- We assume of course that `bird` has such a method.

**Note:** Just because the name of the parameter is `bird`, it doesn't mean that it's an instance of class `Bird`. Remember that parameter names are entirely subjective and they do not convey any meaning or have any relation to other objects in our code.

We now want to call `can_fly()` with three different objects, each of one of the three types `Bird`, `Owl`, `Penguin`.

````python
bird = Bird()
owl = Owl()
penguin = Penguin()
````

Now, let's call `can_fly()` with those objects:

````python
print("#---Fly test---#")
can_fly(bird)
can_fly(owl)
can_fly(penguin)
````

**Insights:** We have one single interface (the `can_fly()` function), but the result depends on the object we pass to it.

In other words, the same function returns differen results based on the object as its parameter.

This is **polymorphism!**

### Class `object`

In Python, there is a class which is the root of all classes.

This is the `object` class and it is the super class of all other classes, either built-in or custom-made.

Let's do a real example now. We will define some classes for shapes.
- Parent class: `Shape`
- Children classes: `Circle`, `Rectangle`, `Square`

````python
import math

class Shape(object):
  def __init__(self, color='red'):
    self.color = color
````

While class `Shape` inherits from `object`, it is not necessary, since we already know that every class inherits from it anyway.

**Note:** You don't have to provide this particular inheritance in your assignments/exams.

````python
class Circle(Shape):
  def __init__(self, radius, color='blue'):
    super().__init__(color=color)
    self.radius = radius

  def area(self):
    return math.pi * self.radius**2

class Rectangle(Shape):
  def __init__(self, width=1.0, length=1.0, color='orange'):
    super().__init__(color=color)
    self.width = width
    self.length = length

  def area(self):
    return self.width * self.length

class Square(Shape):
  def __init__(self, side=1.0, color='white'):
    super().__init__(color=color)
    self.side = side

  def area(self):
    return self.side**2
````

Let's now create objects from those classes:

````python
sh = Shape('purple')
print('Color of Shape sh:', sh.color)

ci = Circle(radius=5)
print('Radius of Circle ci:', ci.radius)
print('Color of Circle ci:', ci.color)

re = Rectangle(width=2, length=8, color='yellow')
print('Color of Rectangle re:', re.color)
print(re.area())
````

As we saw, we didn't define a color instance attribute in neither of the children classes.

That's ok: all children classes define implicitly the color attribute via the `super()` call on `Shape`'s constructor.

## Multiple inheritance

In Python, a class can inherit from multiple classes.

This is called **multiple inheritance**.

It is very commonly used in software development.

Let's see how it works with a simple (pseudo)-example:

```python
# parent class 1
class Main_1:
  #...

# parent class 2
class Main_2:
  #...

# child inherits from two classes
class SubClass(Main_1, Main_2):
  #...
```

This is how we define multiple inheritance, by just putting the parent class names into the parentheses and separated by commas.

Let's continue with a more realistic example:

````python
class Clock:
  def __init__(self, hours, minutes, seconds):
    """
      All attributes start with '__', thus they are private
      No one can access them from outside the class
    """
    self.__hours = hours
    self.__minutes = minutes
    self.__seconds = seconds

  def set_clock(self, hours, minutes, seconds):
    self.__hours = hours
    self.__minutes = minutes
    self.__seconds = seconds

  def time(self):
    return f'{self.__hours}:{self.__minutes}:{self.__seconds}'

class Calendar:
  def __init__(self, d, m, y):
    self.set_calendar(d, m, y)

  def set_calendar(self, d, m, y):
    self.__d = d
    self.__m = m
    self.__y = y

  def date(self):
    return f'{self.__d}/{self.__m}/{self.__y}'
````

**Note:** See carefully in class `Calendar` where we follow a different pattern for creating instance attributes by calling `set_calendar()` method and passing the parameters with it. Also, since `set_calendar()` is a class attribute, we need to call it using the `self.<attribute_name>` format. **Keep that in mind!**

Now that we defined the two parent classes, let's create a new class which inherits from them:

````python
class CalendarClock(Clock, Calendar):
  def __init__(self, day, month, year, hours, minutes, seconds):
    # call the super classes init methods
    Clock.__init__(self, hours, minutes, seconds)
    Calendar.__init__(self, day, month, year)
````

The child class's constructor, takes all the necessary data as parameters. Then, it calls the constructors of both parent classes. For the latter, we have two options:
- Call them explicitly using the class names (what we did here).
  * This is the most appropriate choice here, since each of the parent classes take different parameters (even though they are equal in number).
- Call them implicitly using `super()`. In this case, we don't need to make two calls (or as many as the parent classes are); `super()` will take care of all of them.
  * **However, be careful!** This will not work in the example above. That is because each parent class takes different parameters; a single `super()` call will take into consideration only the first parent (the left-most one in the parentheses, `Clock` here).
  * Therefore, as a general advise for this class, if you come across with a task requiring multiple inheritance, use the explicit call to the parent constructors.

Let's now create an object of class `CalendarClock`:


````python
calendar_clock = CalendarClock(8, 12, 2023, 16, 45, 7)
print(calendar_clock)
print(calendar_clock.time())
print(calendar_clock.date())
````

As we see, since `CalendarClock` inherits from both `Calendar` and `Clock`, it gets everything they have; hence, we can call the `calendar_clock.time()` and `calendar_clock.date()` to get current time and date, respectively.

### Method Resolution Order (MRO)

There is one more topic that we will cover on inheritance: **Multi-Level inheritance**

In Multi-Level inheritance, the parents also have parents, and their parents have parents, and so on.

There is a *deep inheritance* structure which we call multi-level.
- Think of it as `GrandParent -> Parent -> Child`

Let's define some classes to understand the concept of Multi-Level Inheritance:
```python
class GrandParent(object):
  pass

class Parent(GrandParent):
  pass

class Child(Parent):
  pass
```

Let's now see a more real-world example by revisiting the `Clock` and `Calendar` classes, but this time giving them a multi-level inheritance. We will include a class `SandWatch` as the parent of `Clock` class.

The inheritance structure will be:
- `object -> SandWatch -> Clock`
- `object -> Calendar`
- `Clock & Calendar -> CalendarClock`

````python
class SandWatch(object):
  def __init__(self):
    self.start_sand_watch()
  def start_sand_watch(self):
    print('Sand Watch started')

class Clock(SandWatch):
  def __init__(self, hours, minutes, seconds):
    """
      All attributes start with '__', thus they are private
      No one can access them from outside the class
    """
    super().__init__()
    self.set_clock(hours, minutes, seconds)

  def set_clock(self, hours, minutes, seconds):
    self.__hours = hours
    self.__minutes = minutes
    self.__seconds = seconds

  def time(self):
    return f'{self.__hours}:{self.__minutes}:{self.__hours}'

  # override start_sand_watch
  def start_sand_watch(self):
    print('Sand Watch started in Clock class')



class Calendar(object):
  def __init__(self, d, m, y):
    self.set_calendar(d, m, y)

  def set_calendar(self, d, m, y):
    self.__d = d
    self.__m = m
    self.__y = y

  def date(self):
    return f'{self.__d}/{self.__m}/{self.__y}'


class CalendarClock(Clock, Calendar):
  def __init__(self, day, month, year, hours, minutes, seconds):
    # call the super classes init methods
    Clock.__init__(self, hours, minutes, seconds)
    Calendar.__init__(self, day, month, year)
````

Class `Calendar` is the same as before, the only difference is that it now inherits from `object`. Which is no big deal though, since every class inherits from it either implicitly or explicitly.

Class `CalendarClock` is also the same as before.

Now that we have all classes defined, let's create a `CalendarClock` object:

````python
calendar_clock = CalendarClock(12, 1, 2023, 13, 25, 48)
````

Why did we get this printout?

Here are the steps:
1. The `Clock` constructor calls `super().__init__()`, which fires the `__init__()` method of the `SandWatch`.
2. In the `__init__()` method of the `SandWatch`, we call the `self.start_sand_watch()` method.
3. The `start_sand_watch()` method is overridden inside the `Clock` class. The interpreter calls that message, because it is in the "nearest" ancestor.

**Note:** When we have the same method in different classes in the same inheritance chain, the Python interpreter will execute the *nearest* one.

This rule is called **Method Resoltion Order (MRO)**, which is extremely important in the presence of multi-level inheritance.
- We should always keep it in mind when we call a method that is implemented on different levels.

Python provides a special method to get the MRO of a class:
```python
ClassName.__mro__
```
It returns a tuple of orders of classes that will execute first.

````python
for m in CalendarClock.__mro__:
  print(m)
````

The list means that, for the `CalendarClock` class, the interpreter will look for a method:

a. First in the class itself: `CalendarClock`

b. Then in the first ancestor (i.e., the parent): `Clock`

c. Then in the grandparent: `SandWatch`

b. In the first ancestor which is the parent: `Calendar` (don't get confused of 'b' used here, it's not a mistake; it's because `Clock` and `Calendar` are at the same level).

d. Finally in the root class: `object`

# Class attributes

**Class attributes** are defined at class level, which means they are not nested/defined in any method.

Their values are shared between all instances of a class.

```python

class Rabbit(Animal):
    tag = 1     # this is a class attribute

    def __init__(self, age, parent1=None, parent2=None):
        Animal.__init__(self, age)
        self.parent1 = parent1
        self.parent2 = parent2
        # self.rid: instance variable (=rabbit ID)
        # Rabbit.tag: access class variable
        self.rid = Rabbit.tag
        # incrementing class variable changes it for all
        # instances that may refer to it
        Rabbit.tag += 1
```
- Class attribute `tag` is used here to assign a **unique id** to each new rabbit instance.
- Every time a new `Rabbit` object is created, the `Rabbit.tag` attribute will be equal to its value as it was modified from the previous object.
  * It **will not** reset back to $1$.


Let's run some examples below to observe that:

````python
class Animal(object):     # class object implements basic operations in Python
    def __init__(self, age):
        self.age = age
        self.name = None

    def get_age(self):
        return self.age

    def get_name(self):
        return self.name

    def set_age(self, newage):
        self.age = newage

    def set_name(self, newname=""):
        self.name = newname

    def __str__(self):
        return "animal:" + str(self.name) + ":" + str(self.age)

class Rabbit(Animal):
    tag = 1                       # this is a class variable, initiated outside __init__ or any other method

    def __init__(self, age, parent1=None, parent2=None):
        Animal.__init__(self, age)
        self.parent1 = parent1
        self.parent2 = parent2
        self.rid = Rabbit.tag     # self.rid: instance variable (=rabbit ID), Rabbit.tag: access class variable
        Rabbit.tag += 1           # incrementing class variable changes it for all instances that may reference it

a = Animal(3)
a.set_name("fluffy")

b = Rabbit(4)
print(b.tag, b.rid)

c = Rabbit(4)
print(c.tag, c.rid)
````

# Abstract classes

Abstraction is the concept in OOP that is used to **hide the internal functionality of the classes from the users**.

Abstraction is implemented using *abstract classes*.
- An abstract class in Python is typically created to declare a set of methods that must be created in any child class built on top of this abstract class.
- Similarly, an abstract method is one that doesn't have any implementation.

Abstract classes are classes that **cannot** be instantiated.
- We can't create objects of those classes.

### Types of Methods in Python:

Based on the implementation the methods can be divided into two types:

1. Implemented methods.
2. Un-implemented method.

**Implemented method:** A method which has a both method name and method body is an implemented method. They are also called concrete methods or non-abstract methods. The methods which we have seen so far are implemented, i.e having method name and body as well.

**Un-implemented method:** A method which has only method name and no method body (i.e., `pass`) is an unimplemented method. They are also called as non-concrete or abstract methods.

Their purpose is to provide a *blueprint* of a custom-made class and make code sharing easy among closely-related classes.

Benefits from using abstract classes:
- Avoid code duplication.
- Ensure consistency in how others implement subclasses.
- Ensure everyone remembers how to implement critical methods and properties in subclasses.

### How to declare an abstract method in Python:
Abstract methods are declared by using `@abstractmethod ` decorator:
```python
class Demo:
  @abstractmethod
  def one(self):
    pass
  def two(self):
    print('Implemented method')
```

In the above example, method `one` is abstract (un-implemented), whereas method `two` is not (implemented).

**Comments on abstract methods:**
- By using `@abstractmethod` decorator we can declare a method as an abstract method.
- `@abstractmethod` decorator presents in `abc` module. We should **import** the `abc` module in order to use the decorator.
- Since an abstract method is an unimplemented method, we need to put a `pass` statement, else it will result in error.
- A class which contains abstract methods is called an **abstract class**.
- For abstract methods, implementation **must** be provided in the subclass of abstract class.

## `ABC` module

To create an abstract class, you need the module `abc`:

```python
from abc import ABC, abstractmethod

class AbstractClass(ABC):
    @abstractmethod
    def abstract_method(self):
        pass
```

### Abstract Classes in Python:

- Every abstract class in Python should be derived from the `ABC` class which is present in the `abc` module. Abstract classes can contain constructors, parameters, abstract methods, non-abstract methods.
- Abstract methods should be implemented in the subclass or child class of the abstract class.
- If in subclass the implementation of the abstract method is not provided, then that subclass, automatically, will become an abstract class.
- Then, if any class is inheriting this subclass, then that subclass should provide the implementation for abstract methods.
- Object creation is **not possible** for abstract class.
- We can create objects for child classes of abstract classes to access implemented methods.

### Example: Payroll program

Suppose you need to create a payroll program for the company you are working on.

There are two types of employees: fixed-time and hourly
- fixed-time are paid a fixed monthly salary
- hourly are paid by the hour

Let's assume we came up with four classes to develop for creating the program:
* `Employee`
* `FullTimeEmployee`
* `HourlyEmployee`
* `Payroll`

Inheritance:
- `Employee -> FullTimeEmployee`
- `Employee -> HourlyEmployee`
- `Payroll` (this is an independent class from the other ones).

#### Employee class

This will be the abstract class, from where `FullTimeEmployee` and `HourlyEmployee` will inherit attributes and methods.

Abstract class `Employee` will have an abstract method that calculates the employee's salary
- we expect a method of the same name to exist on every subclass of the parent abstract class

```python
from abc import ABC, abstractmethod


class Employee(ABC):
    def __init__(self, first_name, last_name):
        self.first_name = first_name
        self.last_name = last_name

    @abstractmethod
    def get_salary(self):
        pass
```

#### FullTimeEmployee class

Inherits from `Employee` class.

Provides implementation for `get_salary` method.

```python
class FulltimeEmployee(Employee):
    def __init__(self, first_name, last_name, salary):
        super().__init__(first_name, last_name)
        self.salary = salary

    def get_salary(self):
        return self.salary
```

#### HourlyEmployee class

Also inherits from `Employee` class.

Differs from `FullTimeEmployee` in that it requires attributes for rate (hourly compensation), and hours worked at any month.

```python
class HourlyEmployee(Employee):
    def __init__(self, first_name, last_name, worked_hours, rate):
        super().__init__(first_name, last_name)
        self.worked_hours = worked_hours
        self.rate = rate

    def get_salary(self):
        return self.worked_hours * self.rate
```

#### Payroll class

The last class is the `Payroll`.

Adds an employee in a list and returns their salary.

````python
from abc import ABC, abstractmethod


class Employee(ABC):
    def __init__(self, first_name, last_name):
        self.first_name = first_name
        self.last_name = last_name

    @abstractmethod
    def get_salary(self):
        pass
````

**Remember!** An abstract class is not instantiated, therefore you cannot create an object out of it.

````python
empl = Employee('John', 'Smith')
````

````python
class FulltimeEmployee(Employee):
    def __init__(self, first_name, last_name, salary):
        super().__init__(first_name, last_name)
        self.salary = salary

    def get_salary(self):
        return self.salary

class HourlyEmployee(Employee):
    def __init__(self, first_name, last_name, worked_hours, rate):
        super().__init__(first_name, last_name)
        self.worked_hours = worked_hours
        self.rate = rate

    def get_salary(self):
        return self.worked_hours * self.rate

class Payroll:
    def __init__(self):
        self.employee_list = []

    def add(self, employee):
        self.employee_list.append(employee)

    def printt(self):
        for e in self.employee_list:
            print(f"{e.last_name}, {e.first_name} \t ${e.get_salary()}")
````

````python
payroll = Payroll()

payroll.add(FulltimeEmployee('John', 'Doe', 6000))
payroll.add(FulltimeEmployee('Jane', 'Doe', 6500))
payroll.add(HourlyEmployee('John', 'Clark', 50, 15))

payroll.printt()
````
