```python
sales = float(input("Enter sales: "))
cogs = 0.4 * sales
gross_profit = sales - cogs

print(f"Sales: ${sales:.2f}")
print(f"COGS: ${cogs:.2f}")
print(f"Gross Profit: ${gross_profit:.2f}")

tax_rate = float(input("Enter tax rate in percent: "))
taxes = gross_profit * (tax_rate / 100)
net_income = gross_profit - taxes

print(f"Taxes: ${taxes:.2f}")
print(f"Net Income: ${net_income:.2f}")
```

```python
num1 = float(input("Enter first number: "))
op = input("Enter operator (+, -, *, /): ").strip()
num2 = float(input("Enter second number: "))

if op == "+":
    result = num1 + num2
    print("Result:", result)
elif op == "-":
    result = num1 - num2
    print("Result:", result)
elif op == "*":
    result = num1 * num2
    print("Result:", result)
elif op == "/":
    if num2 == 0:
        print("Error: Division by zero.")
    else:
        result = num1 / num2
        print("Result:", result)
else:
    print("Error: Unknown operator.")
```

```python
import math

TICKET_PRICE = 10
BASE_ATTENDEES = 20
FIXED_COST = 200

print("Ads\tProfit")

ads = 0
while ads <= 200:
    additional = 2 * math.sqrt(ads)
    attendees = BASE_ATTENDEES + additional
    revenue = TICKET_PRICE * attendees
    profit = revenue - (FIXED_COST + ads)
    print(f"{ads:.1f}\t{profit:.1f}")
    ads += 25
```

```python
import random

def play_game():
    choices = ["rock", "paper", "scissors"]

    print("\nWelcome to Rock, Paper, Scissors!")

    play = input("Do you want to play a game (yes/no): ").strip().lower()

    while play == "yes":
        user_choice = input("Enter rock, paper, or scissors: ").strip().lower()

        if user_choice not in choices:
            print("Invalid Choice. Try again.")
            # IMPORTANT: don’t 'continue' forever without re-asking play
            # just re-ask user_choice
            continue

        computer_choice = random.choice(choices)
        print("Computer choose", computer_choice)

        if user_choice == computer_choice:
            print("Its a tie")
        elif (user_choice == "rock" and computer_choice == "scissors") or \
             (user_choice == "paper" and computer_choice == "rock") or \
             (user_choice == "scissors" and computer_choice == "paper"):
            print("You win")
        else:
            print("Computer wins")

        play = input("Do you want to play again? (yes/no): ").strip().lower()

    print("Thanks for playing")

play_game()
```

```python
def quicksort(L, reverse=False):
    if len(L) < 2:
        return L[:]

    pivot = L[len(L) // 2]
    less, equal, greater = [], [], []

    for item in L:
        if item < pivot:
            less.append(item)
        elif item == pivot:
            equal.append(item)
        else:
            greater.append(item)

    if not reverse:
        return quicksort(less) + equal + quicksort(greater)
    else:
        return quicksort(greater, reverse=True) + equal + quicksort(less, reverse=True)

print("Quicksort desc:", quicksort([8, 0, 2, 6, 1, 7, 3], reverse=True))
```

```python
def divide(a, b):
    try:
        return a / b
    except TypeError:
        print("Incompatible types")
    except ZeroDivisionError:
        print("b cannot be equal to zero")
    except Exception as e:
        print("unknown error:", e)

print("divide(10, 2) =", divide(10, 2))
print("divide('3', 2) =", divide("3", 2))
```

```python
import math

def getValues(n: int) -> list:
    items = []
    for _ in range(n):
        try:
            item = float(input("Enter a weight value: "))
        except Exception as e:
            print(e)
            item = float("nan")
        items.append(item)
    return items

def calcAverage(alist: list) -> float:
    clean = [x for x in alist if isinstance(x, (int, float)) and not math.isnan(x)]
    if not clean:
        return float("nan")
    return sum(clean) / len(clean)
```

```python
def roi(cost: float, revenue: float) -> float:
    if not isinstance(cost, (int, float)) or not isinstance(revenue, (int, float)):
        raise TypeError("Cost and revenue must be numerical")
    if cost <= 0:
        raise ValueError("Cost must be positive and nonzero")
    return (revenue - cost) / cost
```

```python
class Loan:
    def __init__(self, n: str, a: float, m: int, ir: float) -> None:
        self.name = n
        self.amount = float(a)
        self.maturity = int(m)
        self.ir = float(ir)
        self.org = "N/A"

    def amountIncrease(self, years: int) -> float:
        return self.amount * (1 + self.ir / 100) ** years

    def increaseDifference(self, other: "Loan", years: int) -> float:
        return self.amountIncrease(years) - other.amountIncrease(years)

mortgage = Loan("mortgage", 100000.0, 30, 3)
refinanced_mortgage = Loan("refinanced", 100000.0, 15, 2)

print("Loan diff after 10 years =", mortgage.increaseDifference(refinanced_mortgage, 10))
```

```python
class Greeting:
    def __init__(self, first, last):
        self.first = first
        self.last = last

    def __str__(self):
        return "Instance of class Greeting"

    def display(self, formal=True):
        if formal:
            print(f"{self.last}, {self.first}")
        else:
            print(self.first)

name = Greeting("John", "Doe")
print(name)
name.display()
name.display(formal=False)
```

```python
class Student:
    def __init__(self, name, number):
        self.__name = name
        self.__number = number
        self.__courses = []

    def get_name(self):
        return self.__name

    def get_courses(self):
        return self.__courses

    def set_name(self, newname):
        self.__name = newname

    def __len__(self):
        return len(self.__courses)

    def enroll(self, course):
        if course in self.__courses:
            print("You have already enrolled in", course)
        else:
            self.__courses.append(course)

s = Student("Bob", 12345)
s.enroll("ABC123")
s.enroll("ABC123")
s.enroll("ABC567")
print("Courses:", s.get_courses())
print("Num courses:", len(s))
```

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def distance(self, other):
        return ((self.x - other.x) ** 2 + (self.y - other.y) ** 2) ** 0.5

class Rectangle:
    def __init__(self):
        self.p1 = Point(0, 0)
        self.p2 = Point(4, 0)
        self.p3 = Point(4, 4)
        self.p4 = Point(0, 4)
        self.width = self.p1.distance(self.p2)
        self.height = self.p2.distance(self.p3)

    def area(self):
        return self.width * self.height

r = Rectangle()
print("Rectangle area:", r.area())
```
