```python
sales = float(input("Enter sales: "))
cogs = 0.4 * sales
gross_profit = sales - cogs
print(f'Sales: ${sales:.2f}')
print(f"COGS: ${cogs:.2f}")
print(f"Gross Profit: ${gross_profit:.2f}")
tax_rate = float(input("Enter tax rate in percent: "))
taxes = gross_profit * (tax_rate / 100)
net_income = gross_profit - taxes
print(f"Taxes: ${taxes:.2f}")
print(f"Net Income: ${net_income:.2f}") 
```

```python
num1 = float(input("Enter first number: "))
op = input("Enter operator (+, -, *, /): ").strip()
num2 = float(input("Enter second number: "))

if op == "+":
      result = num1 + num2
      print("Result:", result)
elif op == "-":
      result = num1 - num2
      print("Result:", result)
elif op == "*":
      result = num1 * num2
      print("Result:", result)
elif op == "/":
    if num2 == 0:
        print("Error: Division by zero.")
    else:
        result = num1 / num2
        print("Result:", result)
else:
      print("Error: Unknown operator.")
```

```python
import math

TICKET_PRICE = 10
BASE_ATTENDEES = 20
FIXED_COST = 200

print("Ads\tProfit")

ads = 0
while ads <= 200:
    additional = 2 * math.sqrt(ads)
    attendees = BASE_ATTENDEES + additional
    revenue = TICKET_PRICE * attendees
    profit = revenue - (FIXED_COST + ads)
    print(f"{ads:.1f}\t{profit:.1f}")
    ads += 25
```

```python
import random

def play_game():
    choices = ["rock", "paper", "scissors"]

    print("Welcome to Rock, Paper, Scissors!")

    play = input("Do you want to play a game:").lower()

    while play == "yes":
        
        user_choice = input("Enter rock, paper, or scissors:").lower()

        if user_choice not in choices:
            print("Invalid Choice. Try again")
            continue
        
        computer_choice = random.choice(choices)

        print("Computer choose", computer_choice)

        if user_choice == computer_choice:
            print("Its a tie")

        elif (user_choice == "rock" and computer_choice == "scissors") or \
             (user_choice == "paper" and computer_choice == "rock") or \
             (user_choice == "scissors" and computer_choice == "paper"):
             print("You win")
        else:
            print("Computer wins")
        
        play = input("Do you want to play again? (yes/no):").lower()

    print("Thanks for playing")

play_game()
```
